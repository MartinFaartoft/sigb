```python
#convenience methods

import numpy
import cv2
from pylab import *
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.path as path

def show_image(img):
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    if len(img.shape)<3:
        plt.gray()
    plt.axis("off")
    plt.show()

def showFigures(**imgs):
    plt.figure()
    plt.axis("off")
    for (counter, (k,v)) in enumerate(imgs.items()):
        plt.subplot(1,len(imgs), counter)
        if len(v.shape)<3:
            plt.gray()
        plt.axis("off")
        plt.imshow(v)
        plt.title(k)
    show()
    
def displayFunc(vector):
    points=[]
    for i in range(len(vector)):
        points.append((i,vector[i]))
    bins=range(256)
    n=vector
    plt.grid(None, 'major', 'both')
    plot(bins, n, 'k-', linewidth = 5)
    axis([-2, 256, -2, 256])
    plt.show()
    
def displayHistogram(n, bins, title=""):
    fig = plt.figure()
    plt.title("Histogram - " + title)
    ax = fig.add_subplot(111)
    
    # get the corners of the rectangles for the histogram
    left = np.array(bins[:-1])
    right = np.array(bins[1:])
    bottom = np.zeros(len(left))
    top = bottom + n
    
    
    # we need a (numrects x numsides x 2) numpy array for the path helper
    # function to build a compound path
    XY = np.array([[left,left,right,right], [bottom,top,top,bottom]]).T
    
    # get the Path object
    barpath = path.Path.make_compound_path_from_polys(XY)
    
    # make a patch out of it
    patch = patches.PathPatch(barpath, facecolor='blue', edgecolor='gray', alpha=0.8)
    ax.add_patch(patch)
    
    # update the view limits
    ax.set_xlim(left[0], right[-1])
    ax.set_ylim(bottom.min(), top.max())
    
    plt.show()
```

```python
#black out grayscale pixels below a certain threshold

def black_out_below_threshold(matrix, threshold):
    out = copy(matrix)
    above_threshold_indices = out < threshold
    out[above_threshold_indices] = 0
    return out

I = cv2.imread("exercises1/Hello.jpg")
I = cv2.cvtColor(I, cv2.COLOR_BGR2GRAY)
I = black_out_below_threshold(I, 150)
show_image(I)
```

```python
#applying filters to grayscale images

def apply_filter(image, func):
    for index, value in ndenumerate(image):
        image[index] = func(value)

img = cv2.imread("exercises1/Hello.jpg")
img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

inverter = range(256)
inverter.reverse()
apply_filter(img, lambda x: inverter[x])

show_image(img)
```

```python
#histogram equalization (increases contrast and enhances image)

img = cv2.imread("exercises3/Eye1.jpg") 
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
img_eq = cv2.equalizeHist(img)
showFigures(Before=img, After=img_eq)
```

```python
#pick a single channel, use histogram equalization on that, merge and display

def create_and_display_histogram(array, title=""):
    n,bins = histogram(array, 256, normed=True)
    displayHistogram(n, bins, title)

img = cv2.imread("exercises3/colorProblem.jpg")
b,g,r = cv2.split(img)

#equalize blue
b_eq = cv2.equalizeHist(b)

create_and_display_histogram(b, "Blue")
create_and_display_histogram(b_eq, "Equalized blue")
create_and_display_histogram(g, "Green")
create_and_display_histogram(r, "Red")

#merge the channels to a new image
img_eq = cv2.merge([b_eq, g, r])

showFigures(Source=img, Blue_Equalized=img_eq)
```
