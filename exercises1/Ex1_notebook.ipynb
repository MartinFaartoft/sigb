## Example 1

Run the cell bellow and see how you can load and show and image in python. 
It will opens up a python window and displays the image "Hello.jpg". You can close the window by pressing the space key. 

```python
import cv2
img = cv2.imread("Hello.jpg")

cv2.namedWindow("Picture")
cv2.imshow("Picture", img)
while True:
    ch = cv2.waitKey(10)
    if ch == 32:# Break by SPACE key
        break
cv2.destroyWindow("Picture")
```

## Example 2

This example loads a video file and shows the video in a python window. 

```python
cap = cv2.VideoCapture("video1.mp4")
cv2.namedWindow("input")
f=True
while(f):
    f, img = cap.read()
    if f==True:
        cv2.imshow("input", img)
        ch = cv2.waitKey(1)
        if ch == 32:# Break by SPACE key
            break
cv2.destroyWindow("input")
```

## Example 3

This example capture the image from your camera (assuming that a camera is connected to your computer). 

```python
cv2.namedWindow("camera",1)
capture = cv2.VideoCapture(0)
ret=True
while ret:
    imgs = []
    ret, img =capture.read()
    imgs.append(img)
    cv2.imshow("camera", img)
    ch = cv2.waitKey(10)
    if ch == 32:# Break by SPACE key
        break
capture.release()
cv2.destroyWindow("camera")
```
