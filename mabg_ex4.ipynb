```python
import numpy as np
import cv2
from pylab import *
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import matplotlib.path as path
%matplotlib inline
```

```python
def show_image(img):
    plt.figure(figsize=(10, 10))
    plt.imshow(img)
    if len(img.shape)<3:
        plt.gray()
    plt.axis("off")
    plt.show()
    
def showFigures(**imgs):
    plt.figure()
    plt.axis("off")
    for (counter, (k,v)) in enumerate(imgs.items()):
        plt.subplot(1,len(imgs), counter)
        if len(v.shape)<3:
            plt.gray()
        plt.axis("off")
        plt.imshow(v)
        plt.title(k)
    show()
```

```python
#1
import imp
tools = imp.load_source('tools',"exercises4/Exercises4_Tools.py")
#a
def boxfilter(I,n):
    normFactor = n**2 
    myFilter = np.ones((n, n))/normFactor 
    imFiltered = cv2.filter2D(I, -1, myFilter)
    return imFiltered
#b
I = cv2.imread("exercises4/Lena.png")
lena_gray = cv2.cvtColor(I, cv2.COLOR_RGB2GRAY)

#c
lena_smoothed = boxfilter(lena_gray,3)
#d
showFigures(Orig=lena_gray, lena_5 = boxfilter(lena_gray,5), lena_9 = boxfilter(lena_gray,9),
            diff=cv2.absdiff(lena_gray, boxfilter(lena_gray,3)))
show_image(cv2.absdiff(lena_gray, boxfilter(lena_gray,3)))
#e

kernela = np.matrix('1 0 0; 0 1 0; 0 0 1')
kernelb = np.matrix('-1 0 1; -2 0 2; -1 0 1')
kernelc = np.matrix('-2 -2 0; -2 6 0; 0 0 0')
kerneld = np.matrix('3 10 3; 0 0 0; -3 -10 -3')

i_a = cv2.filter2D(lena_gray, -1, kernela)
i_b = cv2.filter2D(lena_gray, -1, kernelb)
i_c = cv2.filter2D(lena_gray, -1, kernelc)
i_d = cv2.filter2D(lena_gray, -1, kerneld)



showFigures(A=kernela, B= kernelb, C=kernelc, D=kerneld)
showFigures(A=i_a, B=i_b, C=i_c, D=i_d)
```

```python
#2
#a
M,N=lena_gray.shape
noise = np.random.rand(M,N)*60
noise=tools.to_uint8(noise)
LenaGray_uniformNoise = cv2.add(lena_gray , noise)
show_image(LenaGray_uniformNoise)
```

```python
#b
gaussianNoise = np.random.normal(0, 50,(M,N))
gaussianNoise=tools.to_uint8(gaussianNoise)

LenaGray_gaussianNoise = cv2.add(lena_gray , gaussianNoise)
show_image(LenaGray_gaussianNoise)
```

```python
#c
def saltpeppernoise(I,density):
    M,N=I.shape
    copy = np.copy(I)
    random = np.random.rand(M,N)
    copy[(random < density)] = [0]
    copy[(random > 1 - density)] = [255] 
    return copy

#d
LenaGray_SaltPPrNoise = saltpeppernoise(lena_gray, 0.005)
show_image(LenaGray_SaltPPrNoise)
    
```

```python
#e og #f
#LenaGray_uniformNoise , LenaGray_SaltPPrNoise, LenaGray_gaussianNoise

l_sp = boxfilter(LenaGray_SaltPPrNoise, 13)
l_un = boxfilter(LenaGray_uniformNoise, 7)
l_gn = boxfilter(LenaGray_gaussianNoise, 7)

showFigures(SP= l_sp, UN= l_un, GN=l_gn)
```

```python
#g
kern_size = 7

l_gauss_sp = cv2.GaussianBlur(LenaGray_SaltPPrNoise,(kern_size,kern_size),0)

l_gauss_un = cv2.GaussianBlur(LenaGray_uniformNoise,(kern_size,kern_size),0)

l_gauss_gn = cv2.GaussianBlur(LenaGray_gaussianNoise,(kern_size,kern_size),0)

showFigures(SP= l_gauss_sp, UN= l_gauss_un, GN=l_gauss_gn)
```
